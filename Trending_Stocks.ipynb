##### Importing the libraries

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

##### Downloading contents of the web page

```python
response = requests.get('https://finance.yahoo.com/trending-tickers/')
```

##### Checking the status code of the response

```python
response.status_code
```

##### Creaing Soup Object

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

##### Looking for Table with class 'W(100%)' for the required data

```python
table = soup.find('table', class_='W(100%)')
table
```

##### Initializing the required lists

```python
symbol = []
name = []
last_price = []
market_time = []
change = []
percentchange = []
volume = []
market_cap = []
```

##### Collecting Data 

```python
for row in table.tbody.find_all('tr'):
    columns = row.find_all('td')
    
    if (columns != []):
        symbol.append(columns[0].text)
        name.append(columns[1].text)
        last_price.append(columns[2].text)
        market_time.append(columns[3].text)
        change.append(columns[4].text)
        percentchange.append(columns[5].text)
        volume.append(columns[6].text)
        market_cap.append(columns[7].text) 
```

##### Creating the DataFrame

```python
df = pd.DataFrame({'Symbol':symbol,'Name':name,'Last_Price':last_price, 'Market_Time':market_time, 'Change':change, '%Change':percentchange, 'Volume':volume, 'Market_Cap':market_cap})
df
```

##### Exporting the scraped stocks data to excel

```python
df.to_excel('Desktop/trending_stocks.xlsx')
```
